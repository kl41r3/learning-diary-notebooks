`enumerate`的用法

````python
names = ["Alice", "Bob", "Charlie"]

for i, name in enumerate(names):
    print(f"{i}: {name}")
````

等价于

````python
for i in range(len(names)):
    print(f"{i}: {names[i]}")
````

`TypeVar`: Type Variable.

````python
from typing import Generic, TypeVar

T = TypeVar('T')

class Box(Generic[T]):
    def __init__(self, content: T):
        self.content = content

    def get(self) -> T:
        return self.content
````
